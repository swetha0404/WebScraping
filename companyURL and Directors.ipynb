```python
import requests as req
from bs4 import BeautifulSoup
import xlrd
import xlwt
from xlwt import Workbook

#this frunction is used to get the directors name of a company from zauba
def addDirectors(company,link):
    if link!=None:
    
        link=link.replace("company","company-directors")
        r=req.get(link)
        soup=BeautifulSoup(r.content)
        attr={"class":"table table-striped"}
        names=[]
        wb=Workbook()
        sheet=wb.add_sheet('Directors')
        for i in soup.find_all("table",attr):
            for j in i.find_all("tr"):
                for k in j.find_all("td"):
                    for l in k.find_all("h5"):
                        for m in k.find_all("a"):
                            if m.text.isupper():
                                names.append(m.text)
        print(names)
        
#extracts the information of the companies whoses name is there on the excel sheet dowloaded from cibil
def companySearch(s):
    if s!="":
        company=s.replace(" ","-")
        company=company.lower()
        
        url="https://www.zaubacorp.com/companysearchresults/"+company
        attr={"class":"container basic-cont"}
        r=req.get(url)
        soup=BeautifulSoup(r.content)


        l=soup.find_all("div",attr)
        child=list(l[0].children)
        text=str(child[2])
        if text[4:len(text)-4].startswith("We are unable to find"):
            print(s.upper())
            print("Company information not available in Zauba")
            return None
            
        else:
            
            l=soup.find_all("h5")            
            id=str(l[0])[4:len(l[0])-6]         
            url="https://www.zaubacorp.com/company/"+company.upper()+"/"+id
            print(s.upper())
            print(url)
                     
            return url
        
#path of the cibil excel sheet
loc="G:\HS302_PS11\cgSummaryReport_1601003730608.xls"
wb=xlrd.open_workbook(loc)
wb1=Workbook()
sheet=wb.sheet_by_index(0)
sheet1=wb1.add_sheet('URL of Companies')
sheet1.write(0,0,"")
sheet1.write(0,1,"URL of Company")
s_no=1

for i in range(3,sheet.nrows-1):
    #adds the return values from the function into the excel sheet
    l=companySearch(sheet.cell_value(i,1))
    if l!=None:
        sheet1.write(s_no,0,str(s_no))
        sheet1.write(s_no,1,l)
        s_no+=1
    addDirectors(sheet.cell_value(i,1),l)
        
wb1.save("G:/HS302_PS11/URL of Companies.xls")
    
        
   
```
